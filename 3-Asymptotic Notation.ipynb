## Useful Rules
$$g \in o(f) \implies g \in O(f)$$
$$g \in \theta(f) \implies g \in O(f)$$
$$g \in \Omega(f) \implies f \in O(g)$$

## Big O notation

#### Definition:
$$g \in O(f) \iff \exists c>0 \ \exists n_0>0 \ \forall n>n_0 \ g(n) \leq c \cdot f(n)$$

which can be rewritten as
$$g \in O(f) \iff \exists c>0 \ \exists n_0>0 \ \forall n>n_0 \ \frac{g(n)}{f(n)} \leq c$$

explain: for all n greater than a chosen n0, g(n) / f(n) is always less than or equal to a constant c

#### My Understanding of Big-O:
g is upper-bounded by f (not strict bound, can grow faster (little o notation) or same (theta notation))

## Little o notation

#### Definition:
$$g \in o(f) \iff \lim_{n\to\infty} \frac{g(n)}{f(n)}=0$$

#### My Understanding of Little-o:
g is absolutely faster, can't be in the same complexity class (growth) as f.

## Big Omega notation

#### Definition:
$$g \in O(f) \iff \exists c>0 \ \exists n_0>0 \ \forall n>n_0 \ g(n) \geq c \cdot f(n)$$

#### My understanding of big omega:
g is lower bounded by f, (not strict bound, can be slower or same)
